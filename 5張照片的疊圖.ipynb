```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

im5 = mpimg.imread('.JPG')
plt.imshow(im5)
```

```python
b_im5 = im5[:,:,2]
print(b_im5)
print(b_im5.shape)
```

```python
import numpy as np 
b_im1 = im1[:,:,2]
b_im = np.stack((b_im5,b_im1))
print(b_im)
print(b_im.shape)
```

```python
print(np.mean(b_im))
b_ave=np.mean(b_im,axis=0)
plt.imshow(b_ave)
```

```python
im = np.stack((im5,im1,im2,im3,im4),axis=0)
im_ave = np.mean(im,axis=0)
print(im_ave.shape)

plt.imshow(im_ave.astype(np.uint8))
```

```python
print(np.median(b_im))
b_med=np.median(b_im,axis=0)
plt.imshow(b_med)
```

```python
im = np.stack((im5,im1,im2,im3,im4),axis=0)
im_ave = np.mean(im,axis=0)
print(im_ave.shape)

plt.imshow(im_ave.astype(np.uint8))
plt.axis("off")
plt.savefig("black rabbit",bbox_inches='tight',pad_inches=0)
```
